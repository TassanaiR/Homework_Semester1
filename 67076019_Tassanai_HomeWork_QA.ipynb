````python
import os
from openai import OpenAI

#openai_client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
#model_name = "gpt-3.5-turbo"

openai_client = OpenAI(api_key=os.environ.get("OPENTYPHOON_API_KEY"),base_url="https://api.opentyphoon.ai/v1")
model_name = "typhoon-v1.5x-70b-instruct"
````

````python
system_prompt = """You are a helpful assistant that answers \
questions truthfully and factually. Your answers should be \
short and concise. If you don't know the answer, say I am sorry boss.
"""
````

````python
def complete(user_prompt):
        completion = openai_client.chat.completions.create(
        model=model_name,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt},
        ],
        temperature=0,
        max_tokens=100,
        )
        return completion
````

````python
def answer(question, context):
    prompt = f"""Please answer the following question:

    Question:

    ```{question}```

    Use the following context to find the answer:

    ```{context}```
    """

    completion = complete(prompt)
    content = completion.choices[0].message.content
    if content == None :
      print("Can't answer")
    else :
      print(content)
      #return content
````

````python
dog = """###สุนัขสายพันธุ์คอร์กี้ อาหาร วิธีเลี้ยง และราคา

**ข้อมูลทั่วไปสำหรับสุนัขพันธุ์คอร์กี้**

คอร์กี้ (Corgi) สุนัขสายพันธุ์ทรงเลี้ยงของควีนเอลิซาเบธที่ 2 ที่ได้มัดใจทาสหมาทั้งโลก ด้วยหน้าแหลมยาวเหมือนสุนัขจิ้งจอก แต่มาพร้อมขาที่แสนป้อมเตี้ย และหางสั้น ที่ทำให้ท่าเดินของพวกมันดูดุ๊กดิ๊ก เหมือนส่ายก้นอยู่ตลอดเวลา จนได้ฉายาว่า “หมาเตี้ย” และด้วยความนิยมของเจ้าหมาคอร์กี้ที่เพิ่มขึ้นในประเทศไทย เลยมาทำคู่มือการเลี้ยงดูสุนัขสายพันธุ์คอร์กี้ ให้ทาสหมาที่กำลังตัดสินใจเลี้ยงน้อง ๆ กัน ไม่ว่าจะเป็นนิสัย อาหารการกินของคอร์กี้ ราคาในฟาร์มไทย เราก็รวบรวมมาให้หมดแล้ว มาดูกันเลย

**ประวัติของสุนัขพันธุ์คอร์กี้**

หมาคอร์กี้ (Corgi) มีชื่อเต็มว่า เวลช์ คอร์กี้ (Welsh corgi) เนื่องจากถิ่นกำเนิดของพวกมันมาจากแคว้นเวลส์ (Wales) แห่งสหราชอณาจักร (United Kingdom) โดยนักวิชากรสันนิษฐานว่าสายพันธุ์คอร์กี้ปรากฏครั้งแรกในสมัยคริสต์ศตวรรษที่ 10 เพื่อเป็นสุนัขต้อนแกะ และเป็นพี่น้องตระกูลเดียวกับสปิตซ์ (Spitz) โดยคำว่า Cor ในภาษาเวลส์หมายถึง “แคระ” และ Gi มาจาก Ci ที่แปลว่า “สุนัข”
โดยหมาคอร์กี้นั้น ถือเป็นสายพันธุ์ประจำราชวงศ์อังกฤษในขณะนี้เลยก็ว่าได้ โดยจุดเริ่มต้นจาก “ซูซาน”สุนัขทรงโปรดที่ควีนเอลิซาเบธได้รับเป็นของขวัญเนื่องในวันคล้ายวันพระราชสมภพปีที่ 18 จนกระทั่ง 6 ทรรศวรรษต่อมา ควีนเอลิซาเบธที่ 2 ทรงเลี้ยงสุนัขคอร์กี้กว่า 30 ตัวไว้ในวังเลยทีเดียว
หมาคอร์กี้ได้รับการขึ้นทะเบียนสายพันธุ์ครั้งแรกโดย The Kennel Club ที่ประเทศอังกฤษในช่วงยุค 1920 และต่อมามีการแตกแยกเป็นอีกสองสายพันธุ์ย่อยชื่อ The Pembroke Welsh Corgi (พ็อมโบรค เวล์ช คอร์กี้) และ The Cardigan Welsh Corgi (คาร์ดิแกน เวล์ช คอร์กี้) โดยสายพันธุ์ที่เป็นที่นิยมในประเทศไทยคือ The Pembroke Welsh Corgi (พ็อมโบรค เวล์ช คอร์กี้) นั่นเอง

**ลักษณะนิสัยสุนัขพันธุ์คอร์กี้**

แม้ลักษณะภายนอกของหมาคอร์กี้ดูตัวเล็ก ดุ๊กดิ๊ก แท้จริงแล้วพวกมันใจใหญ่ ไม่กลัวคนแต่ไม่ก้าวร้าว แถมยังกล้าหาญ และกะตือรือร้นมาก ๆ ไม่ต่างจากสุนัขพันธุ์ใหญ่อื่น ๆ เพียงแต่รูปร่างแสนป้อมทำให้ท่าทางการแสดงออกของคอร์กี้ดูเบาและน่ารักขึ้นเท่านั้นเอง ยิ่งจุดประสงค์ดั้งเดิมของสายพันธุ์คอร์กี้คือสุนัขใช้งานในฟาร์มแกะ ทำให้นิสัยตามสัญชาติญาณของพวกมัน มีความฉลาด รวดเร็ว และมีความอดทนสูงมากเช่นกัน
แน่นอนว่าคอร์กี้เป็นน้องหมาพันธุ์เล็กที่ชอบใช้พลังงานและเข้าสังคมมาก ดังนั้นเจ้าของควรพาไปพบปะสังคม และพาไปทำกิจกรรมใหม่ ๆ และไม่ควรทิ้งให้อยู่บ้านตัวเดียวเป็นเวลานาน เพราะอาจจะทำให้พวกมันเกิดความเครียดและสะสมพฤติกรรมก้าวร้าว

**มาตรฐานสายพันธุ์เเละลักษณะทางร่างกายที่ดีสำหรับสุนัขพันธุ์คอร์กี้**

เมื่อเลือกซื้อหมาคอร์กี้ สิ่งที่คือต้องสังเกตุหลัก ๆ คือ กะโหลกใบหน้า หาง และความสะอาดพื้นฐานทั่วไป สำหรับสายพันธุ์พ็อมโบรค เวล์ช คอร์กี้ ที่ได้รับความนิยมในประเทศไทยนั้น จะมีใบหน้าที่คล้ายสุนัขจิ้งจอก มาพร้อมหูตั้งสูงทรงสามเหลี่ยม และหางสั้นจุด เหมือนโดนตัดหาง ส่วนสายพันธุ์ คาร์ดิแกน เวล์ช คอร์กี้ จะมีใบหน้าคล้ายสุนัขจิ้งจอกเช่นกัน เพียงแต่มีรูปร่างสูงกว่าเล็กน้อย หลังยาว และมีหางที่ยาวเหมือนสุนัขทั่วไป
โดยสีของหมาคอร์กี้ที่ได้รับการยอมรับตามมาตรฐานสายพันธุ์ คือ สีแดง, สีเทาแกมเหลือง, สีดำ, สีแทน, และสีน้ำตาลเข้มออกแดง โดยอาจจะมีสีขาวปนอยู่ประปรายในบางตัว
นอกจากความแตกต่างของทั้งสองสายพันธุ์แล้ว สิ่งที่ทาสหมาทุกคนต้องสังเกตุเมื่อเลือกซื้อลูกสุนัขคอร์กี้คือ
1.ความร่าเริงและพฤติกรรมของลูกหมา แม้ลูกหมาจะดูนอนทั้งวัน ลูกหมาคอร์กี้ทุกตัวโดยธรรมชาติจะมีแววตาที่สดใส ชวนเล่น ไม่กลัวคน ดังนั้นลูกหมาคอร์กี้ที่ดูกลัวคน และแสดงพฤติกรรมก้าวร้าวแปลกตา อาจจะไม่ได้รับการเลี้ยงดูอย่างดีและเสี่ยงต่อภาวะซึมเศร้าได้
2.ดวงตาสดใส ไร้คราบน้ำตา ซึ่งลูกสุนัขที่มีขี้ตา คราบน้ำตา และเกาตาบ่อย ๆ อาจส่อถึงอาการติดเชื้อที่ดวงตาของน้อง ๆ ได้
3.ใบหูและช่องหูแห้งสะอาด ไม่มีขี้หูหรือรอยแผลต่าง ๆ
4.ตรวจสอบความผิดปกติของช่องปาก ถ้าลูกสุนัขมีริมฝีปากผิดปกติ เพดานแข็ง นั่นอาจหมายถึงโรคปากแหว่ง-เพดานโหว่ ในลูกสุนัขได้ และจะต้องนำไปพบแพทย์เพื่อวางแผนการรักษาหรือผ่าตัดเมื่อโตขึ้น

**โรคที่ต้องระวังในสุนัขสายพันธุ์คอร์กี้**

หมาคอร์กี้มีอายุขัยเฉลี่ยอยู่ที่ 12-15 ปีเหมือนสุนัขสายพันธุ์เล็กอื่น ๆ และเนื่องจากรูปร่างป้อมเตี้ยของพวกมันมาจาก สภาวะแคระ (Achondroplasia) ที่สืบทอดตามพันธุกรรมสายพันธุ์ เช่นเดียวกับดัชชุน (Dachshund) ซึ่งแม้สภาวะแคระไม่ใช่โรคประจำตัว แต่ก็ส่งผลให้เกิดปัญหาสุขภาพต่าง ๆ ที่อาจตามมาได้
โรคข้อสะโพกเสื่อม (Hip Dysplasia) ถือเป็นโรคยอดฮิตสำหรับสุนัขแคระอย่างหมาคอร์กี้ ซึ่งเกิดจากความผิดปกติทางพันธุกรรมที่ส่งผลต่อการเจริญพัฒนาของกระดูกข้อสะโพก ทำให้น้อง ๆ เคลื่อนไหว กระโดด ได้ลำบาก เกิดอาการปวดหลังได้ง่าย โดยโรคข้อสะโพกเสื่อมในคอร์กี้สามารถวินิจฉัยพบได้ตั้งแต่เด็ก และการหมั่นออกกำลังกายจะช่วยชะลอความรุนแรงของโรคได้
การเสื่อมของจอประสาทตา PRA (Progressive Retinal Atrophy) เกิดขึ้นได้เมื่อหมาคอร์กี้อายุมากขึ้น ส่งผลให้ความสามารถในการมองเห็นในที่มืดน้อยลง และอาจทำให้บอดสนิทได้ ถึงแม้จะไม่ใช่โรคที่ถึงแก่ชีวิต แต่ก็ส่งผลต่อคุณภาพชีวิตของคอร์กี้ในระยะยาวได้
โรคเลือดไหลไม่หยุด (Von Willebrand's Disease) ซึ่งคล้ายกับโรค โรคฮีโมฟีเลีย (Hemophilia) ของมนุษย์ ซึ่งเกิดจากความผิดปกติทางพันธุกรรมที่ทำให้สุนัขไม่สามารถสร้าง Von Willebrand Factors ซึ่งมีหน้าที่ในเชื่อมให้เกล็ดเลือดเล็ก ๆ มายึดเกาะกันเป็นก้อนก่อนที่จะไปอุดเส้นเลือดที่ฉีดขาด ซึ่งสังเกตุได้เมื่อหมาคอร์กี้เลือดกำเดาไหล หรือเป็นแผลถลอกแล้วเลือดหยุดยากผิดปกติ

**อาหารที่เหมาะสมสำหรับสุนัขสายพันธุ์คอร์กี้**

หมาคอร์กี้สายพันธุ์พ็อมโบรค เวล์ช คอร์กี้ จะมีน้ำหนักเฉลี่ยอยู่ที่ 10 - 13 กิโลกรัม ในขณะที่สายพันธุ์ คาร์ดิแกน เวล์ช คอร์กี้ จะอยู่ที่ 14-17 กิโลกรัม อาหารการกินจึงควรขึ้นอยู่กับน้ำหนักตัว การใช้พลังงาน ช่วงอายุของน้อง ๆ โดยโภชนาการสำคัญในแต่ละช่วงอายุของคอร์กี้ได้แก่

**อาหารสำหรับลูกสุนัขคอร์กี้ 3 เดือนขึ้นไป**

หลังจากหย่านมแม่แล้ว จะเป็นช่วงสำคัญของลูกหมาคอร์กี้ที่ต้องการพลังงานสูงกว่าคอร์กี้ช่วงโตเต็มวัยเสียอีก โดยเจ้าของควรแบ่งอาหารเป็น 3 มื้อในช่วง 3 - 6 เดือนแรก แล้วค่อย ๆ ลดเหลือ 2 มื้อ เมื่ออายุ 8 เดือนขึ้นไป โดยเน้นไปที่สารอาหารเพื่อการเจริญเติบโตของกล้ามเนื้อต่าง ๆ ไม่ว่าจะเป็นแคลเซียม ฟอสฟอรัส วิตามินเอแมกเนเซียม และ โปรตีนอย่างน้อย 22.5% ตามมาตรฐานของ The Association of American Feed Control Officials (AAFCO) เพื่อบำรุงกระดูกและชะลออาการของโรคข้อสะโพกเสื่อม (Hip Dysplasia) ที่เกิดขึ้นบ่อยในสุนัขพันธุ์คอร์กี้ด้วย

**อาหารสำหรับสุนัขคอร์กี้ 12 เดือนขึ้นไป**

หมาคอร์กี้ที่โตเต็มวัยควรกินอาหารในปริมาณ 1 - 1½ ถ้วยต่อวัน โดยอาจจะให้แค่ 1 มื้อ หรือแบ่งเป็น 2 มื้อก็ได้ ตามนิสัยการกินของน้อง ๆ ซึ่งสำหรับน้องคอร์กี้ที่กินจุ ควรแบ่งเป็น 2 มื้อ และควบคุมสารอาหารและปริมาณให้ดี ๆ เนื่องจากสุนัขพันธุ์นี้เสี่ยงต่อสภาวะอ้วนง่ายกว่าสายพันธุ์อื่น และควรเตรียมน้ำสะอาดไว้ตลอดเวลา

**อาหารสำหรับคอร์กี้สูงวัย 7 ปีขึ้นไป**

ถือเป็นช่วงเวลาที่ต้องใส่ใจในน้ำหนักของหมาคอร์กี้เป็นพิเศษ ยิ่งคอร์กี้ตัวไหนที่เป็นโรคข้อสะโพกเสื่อม (Hip Dysplasia) ก็จะมีอาการหนักขึ้นเมื่อแก่ตัวลง ดังนั้นถ้าปล่อยให้น้ำหนักเกิน อาจส่งผลต่อการเคลื่อนไหวและการรับน้ำหนักตัวของน้อง ๆ ได้ และเพิ่มสารอาหารประเภทโปรไบโอติกส์ และพรีไบโอติกส์เพื่อดูแลระบบขับถ่ายของน้อง ๆ ให้สมดุล และหลีกเลี่ยงธัญพืช กระดูก หรือแป้ง ซึ่งอาจส่งผลเสียต่อสุขภาพได้ง่าย

**วิธีการดูแลสุนัขพันธุ์คอร์กี้ ตั้งแต่เป็นลูกสุนัขจนถึงโตเต็มวัย**

**การดูแลลูกสุนัขพันธุ์คอร์กี้ ช่วงอายุ 4-7 เดือน**

ในช่วงเด็ก ๆ ลูกหมาคอร์กี้จะติดนอนทั้งวัน บางครั้งอาจจะใช้เวลา 20 ชั่วโมงต่อวันเลยทีเดียว ซึ่งเป็นเรื่องปกติที่เราจะเพิ่งรับคอร์กี้มาเลี้ยงแล้ววัน ๆ ไม่ได้ทำอะไรนอกจากกินและนอนอย่างเดียว สิ่งสำคัญสำหรับลูกสุนัขคอร์กี้คือ การถ่ายพยาธิและฉีดวัคซีนครั้งแรก ซึ่งสามารถทำได้ตั้งแต่ 6 สัปดาห์ขึ้นไป และคอยสังเกตุอาการแพ้ที่อาจจะเกิดขึ้นได้ในลูกสุนัขบางตัว ไม่ว่าจะเป็นแพ้อาหาร แพ้เกสรดอกไม้ หรืออื่น ๆ ซึ่งให้รีบนำไปพบแพทย์เพื่อวางแผนโภชนาการและการรักษาต่อไป

**การดูแลสุนัขพันธุ์คอร์กี้โตเต็มวัย อายุ 1 ปีขึ้นไป**

สิ่งที่ต้องดูแลหมาคอร์กี้เป็นเรื่องทั่วไป คือ การออกกำลังกาย จากคอร์กี้เป็นสุนัขสายพันธุ์ต้อนแกะ จึงชอบใช้พลังงานมาก เจ้าของควรมีเวลามากพอในการพาน้อง ๆ ไปทำกิจกรรมต่าง ๆ ไม่ว่าจะเป็นเดินเล่น ว่ายน้ำ แต่ไม่ควรออกกำลังกายหนักเกินไปเพราะอาจส่งผลต่อสะโพกและกระดูกของคอร์กี้ได้ ในส่วนของขนและความสะอาด ถึงแม้คอร์กี้จะมีขนสั้น ก็ควรพาไปตัดขนทุก ๆ 1 -2 เดือน และหมั่นแปรงขนบ่อย ๆ เพื่อให้ขนไม่แข็งและดูนุ่มสุขภาพดี

**ราคาหมาพันธุ์คอร์กี้**

เมื่อคุณคิดว่าหมาพันธุ์คอร์กี้ คือหมาพันธุ์ที่ใช่สำหรับคุณแล้ว เรามาดูเรื่องราคากันดีกว่า ราคาสำหรับสุนัขสายพันธุ์คอร์กี้ จะขึ้นอยู่กับสายพันธุ์และแหล่งเพาะพันธุ์พวกมัน ส่วนมากจะอยู่ที่ช่วงราคา 15,000 - 30,000 บาทขึ้นไป สำหรับคนที่ตัดสินใจแล้วว่าต้องการที่จะเลี้ยงสุนัขพันธุ์นี้

**สรุป**

เรียกได้ว่า คอร์กี้ เจ้าสุนัขตัวป้อม หางสั้น เป็นอีกหนึ่งทางเลือกดี ๆ ของทาสหมาหลากหลายไลฟ์สไตล์ ไม่ว่าจะเป็นขนาดที่เล็ก และอุปนิสัยกล้าหาญ ร่าเริง แถมเข้ากับมนุษย์ได้อย่างดีเยี่ยม ซึ่งเรื่องสุขภาพที่ต้องดูแลหลัก ๆ คือเรื่องการออกกำลังกายและการเคลื่อนไหวของน้องที่ต้องไม่หักโหมเกินไป และไม่ทำร้ายช่วงสะโพกของน้อง ๆ โดยการเลือกอาหารที่โปรตีนสูง จะช่วยบำรุงและซ่อมแซมกระดูกและร่างกายของคอร์กี้ได้ดียิ่งขึ้น

"""
IPython.display.Markdown(dog)
````
